```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_bloch_multivector
from qiskit.quantum_info import Statevector
import numpy as np
```

```python
qc = QuantumCircuit(2,2) #Creando circuito. QuantumCircuit(N_Qubits, lineas de medición)
qc.x([0,1]) #Aplicando la compuerta pauli X (o compuerta not en clásica)
qc.draw() #visualizar qubits
sv = Statevector.from_instruction(qc) #crear un vector de estado
sv #imprime de forma matemática el statevector
```

```python
dir(qc) #Estructura de propiedades y métodos de qc
```

```python
plot_bloch_multivector(sv) #imprimir vector de estado
```

```python
qc2 = QuantumCircuit(2) #Se crea un nuevo circuito llamado: "qc2"
#qc2.h(0) #Se aplica compuerta hadamard
qc2.rx(np.pi/6, 0) 
qc2.draw()
sv2 = Statevector(qc2)
plot_bloch_multivector(sv2)
```

```python
#Ahora vamos a crear un estado de Bell
qc3 = QuantumCircuit(3,3)

qc3.h(0)
#qc3.cx(0,1)

qc3.h(1)
#qc3.cx(1,2)

qc3.h(2)

qc3.measure([0,1,2], [0,1,2]) #measure([índice de qubits que voy a medir], [índice de canales clásicos])
qc3.draw()
```

```python
sv3 = Statevector(qc3)
```

```python
plot_bloch_multivector(sv3)
```

```python
from qiskit.visualization import plot_state_qsphere
```

```python
plot_state_qsphere(qc3)
```

```python
from qiskit import Aer
```

```python
Aer.aer.backends()
```

```python
backend = Aer.aer.get_backend("qasm_simulator")
```

```python
job = backend.run(qc3, shots=100000)
result = job.result()
counts = result.get_counts()
```

```python
from qiskit.visualization import plot_histogram
```

```python
plot_histogram(counts)
```

```python
binar = list(result.get_counts().keys())[0]
print(int(binar, 2))
```

```python
from math import pi
qc4 = QuantumCircuit(2,2)

qc4.h(0)
qc4.cx(0,1)
#qc4.x(0)
qc4.rx(pi/2,1)
qc4.measure([0,1], [0,1]) #measure([índice de qubits que voy a medir], [índice de canales clásicos])
qc4.draw()
```
