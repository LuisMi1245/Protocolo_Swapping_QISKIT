```python
import math as m
```

```python
a = m.cos(0.5)
```

```python
print('{:^2}{:^23}{:^23}{:^23}{:^23}{:^23}'.format('i','a','b','y','w','z'))
for i in range(1,21):
    b = m.sin(float(i))
    y = a / b
    w = y * b
    z = a - w
    print('{:^2}{:^23}{:^23}{:^23}{:^23}{:^23}'.format(i,a,b,y,w,z))
```
