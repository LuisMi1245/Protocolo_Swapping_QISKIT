### Cálculo de Control $\xi(t)$ para Sistema de 2 niveles Mediante redes neuronales PINN

El hamiltoniano de mi sistema está dado por:
\begin{equation}
    H(t) = \omega_z \sigma_z + \omega_x \sigma_x + \xi(t)\sigma_{ee} \quad (1)
\end{equation}
con $\omega_{x,z}$ los parámetros del sistema y $\xi(t)$ la función que describe el campo de control desconocido. Además, $\sigma_z = \ket{e}\bra{e}-\ket{g}\bra{g}$, siendo $\ket{e}$ el estado excitado y $\ket{g}$ el estado fundamental.

Con ésto en mente, la dinámica abierta del sistema queda descrita por la ecuación maestra de Markov:
$$\dot{\rho} = -i[H(t), \rho]+ \sum_{i=1,2} \gamma_i(L_i \rho L_i^{\dagger}-(1/2)\{L_i^{\dagger}L_i,\rho\}) \quad (2)$$

En éste caso particular, la disipación se describe por los procesos de absorción ($L_1 = \ket{e}\bra{g}$) y emisión ($L_2 = \ket{g}\bra{e}$) con tasas de $$\gamma_1 = \gamma_{abs} = 0.1$$ $$\gamma_2 = \gamma_{em} = 0.3$$ Además, se configuran los parámetros del sistema $$\omega_z = 2$$ $$\omega_x = 1$$ y la condición inicial $$\xi(0) = 0$$ 

> **Objetivo:** Predecir la función del campo de control $\xi(t)$ óptima que nos permita controlar la dinámica de nuestro sistema, para conducirlo desde un estado inicial a otro estado estacionario de la forma: $\rho_d = (1/2)(\ket{e}\bra{e}+\ket{g}\bra{g})$.

## Parametrización del problema

La ecuación de la dinámica de nuestro sistema (2) puede ser reescrita como una ecuación diferencial lineal de la forma:
$$\frac{dx}{dt} = A(\lambda, u(t))*\mathbf{x}(t) \quad (3)$$ donde $\mathbf{x}(t)$ es un vector real, $\lambda = (\omega_x, \omega_z, \gamma_{abs}, \gamma_{em})$ son los parámetros conocidos del sistema, $\mathbf{u}(t) = \xi(t)$ es el vector unidimensional de control que se quiere hallar.

Para llegar a ésta forma de matrices, debemos escribir la ecuación (2) en términos de matrices:

    1. Partimos definiendo la matriz densidad correspondiente que representa dicho estado de tamaño 2x2:

$$\rho = \begin{pmatrix} \rho_{00} & Re(\rho_{01})+Im(\rho_{01}) \\
                        Re(\rho_{10})-Im(\rho_{10}) & \rho_{11}
\end{pmatrix} = \begin{pmatrix} z_{00} & z_{01r}+z_{01i} \\
                        z_{10r}-z_{10i} & z_{11}
\end{pmatrix} = \rho_{Real} + \rho_{Im}$$

    En tal caso:

$$\rho_{Real} = \begin{pmatrix} \rho_{00} & Re(\rho_{01}) \\
                        Re(\rho_{10}) & \rho_{11}
\end{pmatrix}$$


$$\rho_{Im} = \begin{pmatrix} 0 & Im(\rho_{01}) \\
                        -Im(\rho_{10}) & 0
\end{pmatrix}$$

```python
%%capture
from sympy import init_session
from IPython.display import display, Math
init_session()
ket_g = Matrix([[0],[1]])
ket_e = Matrix([[1],[0]])
```

```python
z00,z11,z01r,z01i  = symbols('z00,z11,z01_r,z01_i',real=True)
rho = Matrix([[z00,z01r+I*z01i],
              [z01r-I*z01i,z11]])
Math(r'\rho'+' = '+latex(rho))
```

    2. Seguidamente, se escribe el hamiltoniano en términos de matrices:

```python
from sympy.physics.matrices import msigma
sigma_z = ket_e*ket_e.T - ket_g*ket_g.T
sigma_x =  ket_e*ket_g.T + ket_g*ket_e.T
sigma_ee = ket_e*ket_e.T
omega_z, omega_x, xi = symbols('omega_z omega_x xi', real=True)
xi = symbols('xi', real=True)
df = symbols('df', cls=Function)
```

```python
H = omega_z*sigma_z + omega_x*sigma_x + xi*sigma_ee
Math(r'H = '+ latex(H))
```

    3. También definimos los términos de disipación:

```python
gamma_1, gamma_2 = symbols('gamma_abs gamma_em', real=True)
L1 = ket_e*ket_g.T
L2 = ket_g*ket_e.T
disip_1 = gamma_1*((L1*rho*L1.T) - (Rational(1/2)*(L1.T*L1*rho + rho*L1.T*L1)) )
disip_2 = gamma_2*((L2*rho*L2.T) - (Rational(1/2)*(L2.T*L2*rho + rho*L2.T*L2)) )
```

```python
(disip_1 + disip_2).subs(gamma_1, 0.1).subs(gamma_2, 0.3)
```

    4. Finalmente obtenemos la ecuación maestra en forma de matriz:

```python
dot_rho = -I*(H*rho - rho*H) + disip_1 + disip_2
Math(r'\dot{\rho} = '+ latex(expand(dot_rho)))
```

```python
simplify(ket_g.T*dot_rho*ket_g)
```

Al describir la matriz $\dot{\rho}$ En términos de la ecuación $\dot{x}=\mathbf{A}\mathbf{x}$, considerando a $\mathbf{x} = (\rho_{gg},\rho_{ee},Re(\rho_{eg}),Im(\rho_{eg}))^{T} = (\rho_{11},\rho_{00},Re(\rho_{01}),Im(\rho_{01}))^{T} $

```python
display(Math(r'\dot{x}_1 = '+latex(simplify(dot_rho[1,1].subs(gamma_1, 0.1).subs(gamma_2, 0.3).subs(omega_z, 2).subs(omega_x, 1)))))
display(Math(r'\dot{x}_2 = '+latex(simplify(dot_rho[0,0].subs(gamma_1, 0.1).subs(gamma_2, 0.3).subs(omega_z, 2).subs(omega_x, 1)))))
display(Math(r'\dot{x}_3 = '+latex(simplify(re(dot_rho[0,1].subs(gamma_1, 0.1).subs(gamma_2, 0.3).subs(omega_z, 2).subs(omega_x, 1))))))
display(Math(r'\dot{x}_4 = '+latex(simplify(im(dot_rho[0,1].subs(gamma_1, 0.1).subs(gamma_2, 0.3).subs(omega_z, 2).subs(omega_x, 1))))))
```

es decir,

$$\dot{x} = \begin{pmatrix} -\gamma_{abs} & \gamma_{em} & 0 & 2\omega_x \\  \gamma_{abs} & -\gamma_{em} & 0 & -2\omega_x\\ 0 & 0 & -\Gamma & 2\omega_z + \xi(t) \\ -\omega_x & \omega_x & -2\omega_z - \xi(t) & -\Gamma\end{pmatrix}\begin{pmatrix}\rho_{11} \\ \rho_{00} \\ Re(\rho_{01}) \\ Im(\rho_{01}))\end{pmatrix}$$ siendo $\Gamma = (1/2)(\gamma_1+\gamma_2)$ la tasa de desfase efectiva.

```python
#Elementos de la diagonal
f1 = -df(z11,t) + simplify(dot_rho[1,1].subs(gamma_1, 0.1).subs(gamma_2, 0.3).subs(omega_z, 2).subs(omega_x, 1))
f2 = -df(z00,t) + simplify(dot_rho[0,0].subs(gamma_1, 0.1).subs(gamma_2, 0.3).subs(omega_z, 2).subs(omega_x, 1))
#Parte Real de los elementos
f3 = -df(z01r,t) + re(simplify(dot_rho[0,1]).subs(gamma_1, 0.1).subs(gamma_2, 0.3).subs(omega_z, 2).subs(omega_x, 1))
#Parte Imaginaria de los elementos
f4 = -df(z01r,t) + im(simplify(dot_rho[0,1]).subs(gamma_1, 0.1).subs(gamma_2, 0.3).subs(omega_z, 2).subs(omega_x, 1))
```

# Red Neuronal informada por la física (PINN)

Importamos el marco de trabajo Pythorch para trabajar nuestra red neuronal de forma dinámica:

```python
import numpy as np
import torch #Marco de trabajo para redes neuronales
from torch import nn #Herramientas para elaborar el modelo de red neuronal
import matplotlib.pyplot as plt
from torch.autograd import grad
import copy
from IPython.display import display, clear_output
import time

dtype=torch.float

#Se establece el hardware para usar en cálculos de la red neuronal 
device = (
    "cuda"
    if torch.cuda.is_available()
    else "mps"
    if torch.backends.mps.is_available()
    else "cpu"
)
print(f"Usando {device} como plataforma")
```

### Función de activación Sin()

Las activaciones no lineales son las que crean las asignaciones complejas entre las entradas y salidas del modelo. En éste caso, dado que Pytorch no posee activaciones tipo Seno, se crea una nueva subclase tipo módulo de Pytorch. Ésta activación se aplica después de transformaciones lineales para introducir la no linealidad. Se ha demostrado que esta elección de activación mejora el rendimiento de los PINN en la resolución de sistemas dinámicos no lineales y ecuaciones diferenciales parciales de alta dimensión (Norambuena et. al 2022).

```python
#Definimos la clase Sin() que hereda de torch.nn.Module
class Sin(nn.Module):
    def __init__(self):
        super().__init__()
    def forward(self, input):
        return torch.sin(input)
```

### Esqueleto de nuestra red neuronal:

La red neuronal consta de:

1. Una capa de entrada de valores de tiempo.
2. 4 capas ocultas, cada una de 200 neuronas.
3. En todas las neuronas de cada capa oculta se ejecutan operaciones lineales y seguido, se aplica la función de activación seno.
4. Una capa de salida que me arroja las funciones $N_i(t)$ de la parametrización.

</center>

```python
#En pytorch, todos los esqueletos de redes neuronales se definen como una clase que heredan los métodos y propiedades de torch.nn.Module
class NeuralNetwork(nn.Module):
    def __init__(self): 
        super().__init__()
        self.sin_f = Sin() 
        self.ns = 200
        self.linear1 = nn.Linear(1, self.ns)
        self.linear2 = nn.Linear(self.ns, self.ns)
        self.linear3 = nn.Linear(self.ns, self.ns)
        self.linear4 = nn.Linear(self.ns, self.ns)
        self.linear_out = nn.Linear(self.ns, 5)
    def forward(self, t):
        #Ésta instrucción asigna a la capa de entrada el vector de entrada "t"
        #Y propaga la salida de cada capa hacia la capa siguiente de forma sequencial
        t = self.sin_f(self.linear1(t))
        t = self.sin_f(self.linear2(t))
        t = self.sin_f(self.linear3(t))
        t = self.sin_f(self.linear4(t))
        t = self.linear_out(t)
        
        z11n = (t[:,0]).reshape(-1,1); 
        z00n = (t[:,1]).reshape(-1,1); 
        z01rn = (t[:,2]).reshape(-1,1); 
        z01in = (t[:,3]).reshape(-1,1); 
        z5n = (t[:,4]).reshape(-1,1); 
        
        return z11n, z00n, z01rn, z01in, z5n
```

### Función del bucle de Entrenamiento

```python
def train_neural_network(input_data, t_grid, learning_rate, batch_number = 1): 
    #Se genera una instancia de nuestro modelo de red neuronal
    modelo = NeuralNetwork()
    model_deep_copy = copy.deepcopy(modelo) #Se hace una copia normal del modelo de red neuronal

    #Se define la función de costo y el optimizador de la función de costo, en éste caso, se usa el optimizador: ADAM.
    optimizador = torch.optim.Adam(modelo.parameters(), lr=learning_rate, betas=[0.999, 0.9999])
    
    #Definimos un contadores de pérdida
    minLoss=1e-6
    Loss_history = []  
    Llim =  1  #Límite de función de costo

    for epoch in range(epochs):    
        print(f"Época: {epoch}")
       
        tiempo_inicio = time.time()  # Registra el tiempo de inicio de la época

        #Se crean mini lotes de tiempo
        t_batch, batch_start, batch_end, batch_size = crear_minibatch(t_grid, t_grid.shape[0], batch_number)
    
        #Se establece el modelo en modo entrenamiento, es decir: pytorch optimiza el proceso de entramiento internamente
        modelo.train()

        loss = 0
        #Realizamos el entrenamiento por cada minibatch (mini lote) de valores de tiempo de tamaño: batch_size
        for batch in range(batch_number):
            #Toma todos los elementos del vector de tiempo que se encuentran desde el índice batch_start hasta batch_end
            t_mb = t_batch[batch_start:batch_end] 
            
            #Se evalúa la red neuronal con el mini lote de valores de tiempo 
            #Y se establecen las soluciones paramétricas de la ecuación maestra
            z11, z00, z01r, z01i, xi = parametric_solutions(t_mb, modelo, input_data)

            # Función de pérdida definida por las Ecs. de Hamilton (simplécticas): Escribiendo explícitamente las Ecs (más rápido)
            Ltot = hamEqs_Loss(t_mb,z11,z00,z01r,z01i, xi, input_data)

            l2_lambda = 0.001
            l2_reg = torch.tensor(0.)
            for param in modelo.parameters():
                l2_reg += torch.norm(param)
            Ltot += l2_lambda * l2_reg

            Ltot.backward(retain_graph=False); #True
            optimizador.step(); 
            loss += Ltot.data.numpy()
       
            optimizador.zero_grad()

            #Ahora tomamos los valores de tiempo del siguiente batch
            batch_start +=batch_size
            batch_end +=batch_size

        Loss_history.append(loss)       

        #Keep the best model (lowest loss) by using a deep copy
        if  epoch > 0.8*epochs and Ltot < Llim:
            #Se crea una copia independiente del modelo con la menor pérdida
            model_deep_copy = copy.deepcopy(modelo)
            Llim=Ltot 
        
        if Ltot < minLoss:
            model_deep_copy =  copy.deepcopy(modelo)
            print('Reach minimum requested loss')
            break
            
        tiempo_fin = time.time()  # Registra el tiempo de finalización de la época

    return Loss_history, model_deep_copy
```

### Función para agregar perturbación estocástica al vector de tiempo

```python
def perturbar_puntos_tiempo(t_0, t_max, t_size=300, sig=0.5):
    grid = torch.linspace(t_0, t_max, t_size).reshape(-1, 1)
    delta_t = grid[1] - grid[0]  
    noise = delta_t * torch.randn_like(grid)*sig
    t = grid + noise
    t.data[2] = torch.ones(1,1)*(-1)
    t.data[t<t_0]=t_0 - t.data[t<t_0]
    t.data[t>t_max]=2*t_max - t.data[t>t_max]
    t.data[0] = torch.ones(1,1)*t_0
    t.requires_grad = False
    return t
```

### Función para crear lotes de datos

```python
def crear_minibatch(t_grid, t_size, batch_number): 
    #Ésta función crea lotes de datos para optimizar el proceso de entrenamiento. No tiene sentido en redes PINN
    batch_size = int(t_size/batch_number)
    batch_start, batch_end = 0, batch_size

    idx = np.random.permutation(t_size)
    t_b = t_grid[idx]
    t_b = t_grid
    t_b.requires_grad = True
    return t_b, batch_start, batch_end, batch_size
```

### Función para evaluar la red neuronal en el vector de tiempo t, y generar soluciones paramétricas

```python
 def parametric_solutions(t, modelo, input_data):
    #Filtramos input_data para obtener solamente las condiciones iniciales de las variables en t=0
    t_0, z00_0, z11_0, z01r_0, z01i_0, xi_0 = input_data[:6]
    
    #Evaluamos nuestra red neuronal en el vector de tiempo t
    N_1, N_2, N_3, N_4, N_5 = modelo(t)
    #N_1, N_2, N_3, N_4 = modelo(t)

     
    #Obtenemos el diferencial entre el vector t y el valor t_0 (ocurre broadcasting)
    dt = t-t_0

    #Se define la función paramétrica, y a su vez, todas las soluciones paramétricas para cada componente del vector x(t) y u(t)
    f = (1-torch.exp(-dt))
    #La mejor función que afecte en menor medida el resultado de la red neuronal
     
    #Soluciones para x(t) (dinámica del sistema)
    Z11 = z11_0 + f*N_1
    Z00 = z00_0 + f*N_2
    Z01r = z01r_0 + f*N_3
    Z01i = z01i_0 + f*N_4

    #Soluciones para u(t) (función de control)
    Z5 = xi_0 + f*N_5
    
    
    return Z11, Z00, Z01r, Z01i, Z5
```

### Función de pérdida definida por las Ecs. de Hamilton (simplécticas):

```python
def hamEqs_Loss(t, z11, z00, z01_r, z01_i, xi, input_data):
    #gamma_abs [6]
    #gamma_em [7]
    #omegax [8]
    #omegaz [9]
    
    # Equations
    f1 = input_data[6]*z11 - input_data[7]*z00 - 2*input_data[8]*z01_i + df(z11, t)
    f2 = -input_data[6]*z11 + input_data[7]*z00 + 2*input_data[8]*z01_i + df(z00, t)
    f3 = input_data[6]*z01_r/2 + input_data[7]*z01_r/2 - input_data[9]*z01_i - z01_i*(input_data[9] + xi) + df(z01_r, t)
    f4 = input_data[6]*z01_i/2 + input_data[7]*z01_i/2 - input_data[8]*z00 + input_data[8]*z11 + input_data[9]*z01_r + z01_r*(input_data[9] + xi) + df(z01_r, t)
    

    # Términos de pérdida del modelo
    L1 = (f1.pow(2)).mean();  
    L2 = (f2.pow(2)).mean(); 
    L3 = (f3.pow(2)).mean(); 
    L4 = (f4.pow(2)).mean(); 
    Lmodel = L1+L2+L3+L4; 

    # Límites para el control (Lcontrol)
    f5 = z00 - 1/2    # rho_00 = 1/2
    f6 = z11 - 1/2    # rho_11 = 1/2
    f7 = z00 + z11 - 1 # Tr(rho) = 1
    #f8 = z01_r # rho_01r = 0
    #f9 = z01_i # rho_01i = 0

    L5 = (f5.pow(2)).mean();  
    L6 = (f6.pow(2)).mean(); 
    L7 = (f7.pow(2)).mean(); 
    #L8 = (f8.pow(2)).mean(); 
    #L9 = (f9.pow(2)).mean(); 
           
    L = Lmodel + 0.1*L7 + 1*(L5 + L6) 

    return L

def df(f,t):
    return grad([f], [t], grad_outputs=torch.ones(t.shape, dtype=dtype), create_graph=True)[0]
```
